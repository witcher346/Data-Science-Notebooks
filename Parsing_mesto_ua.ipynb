```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import os
from IPython.core.display import display, HTML

pd.set_option('display.max_columns', 500)
display(HTML("<style>.container { width:95% !important; }</style>"))
NUM_PAGES = 1500
```

```python
def get_ads(num_page):
    
    urls = 'https://kiev.mesto.ua/sale/?p=' + str(num_page)
    req = requests.get(urls)
    soup = BeautifulSoup(req.text)
    ads = []
    for a in soup.find_all('a', {'class': 'title'}): # find all links to ads
        ads.append(a['href'])
        
    return ads
```

```python
def collect_data(url):
    
    req = requests.get(url)
    soup = BeautifulSoup(req.text)
    ads = []
    
    div = soup.find('div', {'class': 'tab-pane active', 'id': 'info'}) # parsing the table, starting with finding the div of a table
    table = div.table
    properties = {}
    for el in table.find_all('tr'):
        ls = el.find_all('td')
        properties[ls[0].contents[0][:-1]] = ls[1].contents[0] # extracting every table element into a dictionary

    description = soup.find('div', {'class': 'description'})
    properties[description.div.contents[0]] = description.p.contents[0] # extracting description
    properties['url'] = url
    
    return properties
```

```python
data = []

try:
    for num_page in range(1, NUM_PAGES+1):
        print('Collecting from page {}'.format(num_page))
        urls = get_ads(num_page)
        for url in urls:
            data.append(collect_data(url))
            print('Collected from page {}'.format(url))
        print('\n\n')
            
except:
    pass

finally:
    df = pd.DataFrame(data, index=range(0, len(data)))
    df.to_csv(path_or_buf='E:\\mesto_ua.csv')
    print('Dataframe created. Saved as .csv') 
    
```
